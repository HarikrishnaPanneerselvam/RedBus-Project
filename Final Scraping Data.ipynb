```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from datetime import date
from selenium.common.exceptions import *
from selenium.webdriver.common.action_chains import ActionChains
import time
```

```python
# Step 1: Open the Redbus website
driver = webdriver.Chrome()
driver.get("https://www.redbus.in/")

# Step 2-4: Click on each state name button, go to the new page, and get the current URL
state_urls = []

for i in range(10):  # adjust the range to the number of state buttons on the page
    state_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, f"(//div[@class='rtcName'])[{i+1}]"))
    )
    driver.execute_script("arguments[0].click();", state_button)
    new_page_url = driver.current_url
    driver.back()
    state_urls.append(new_page_url)

WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "rtcCards")))
state_cards = driver.find_elements(By.CLASS_NAME, "rtcCards")

all_state_rtc_details = []
for state_card in state_cards:
    state_name_element = state_card.find_element(By.CLASS_NAME, "rtcName")
    WebDriverWait(driver, 10).until(EC.visibility_of(state_name_element))
    state_name = state_name_element.text
    state_rating_element= state_card.find_element(By.CSS_SELECTOR, "div[class*='rating rtcRating']")
    WebDriverWait(driver, 10).until(EC.visibility_of(state_rating_element))
    state_rating = state_rating_element.text
    all_state_rtc_details.append({"State Name": state_name, "State Rating": state_rating, "State URL": new_page_url})

# Step 6: Convert the list details into a Pandas dataframe
df = pd.DataFrame(all_state_rtc_details)

# Step 7: Convert the dataframe into a CSV file
df.to_csv("All_State_Links.csv", index=False)

#Close the browser window
driver.quit()
df
```

```python
# Step 1: Open the required web page
driver = webdriver.Chrome()  # Replace with your preferred browser
driver.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")  

# Step 2: Find the tab buttons
tab_buttons = driver.find_elements(By.CSS_SELECTOR, "div[class^='DC_117_pageTabs']")

# Initialize lists to store route names and links
route_names = []
route_links = []

# Loop through each tab
for tab_button in tab_buttons:
    driver.execute_script("arguments[0].click()", tab_button)
    
    # Step 3: Find the route cards
    route_cards = driver.find_elements(By.CSS_SELECTOR, "div[class*='route_link']")
    
    # Loop through each route card
    for route_card in route_cards:
        # Step 3: Check if route name is available
        route_name_element = route_card.find_element(By.CLASS_NAME, "route")
        if route_name_element:
            # Step 4: Split and store the route name
            route_name = route_name_element.text
            from_to = route_name.split(" to ")
            from_city = from_to[0]
            to_city = from_to[1]
            route_names.append({"From": from_city, "To": to_city})
        else:
            print("Buses not found")
            break
        
        # Step 6: Find the route link
        route_link_element = route_card.find_element(By.CSS_SELECTOR, "a[class*='route']")
        route_links.append(route_link_element.get_attribute("href"))
    
    # Step 9: Check if route card is not available
    if not route_cards:
        print("Not available")

# Step 13: Close the web page
driver.quit()

# Step 14: Store the collected data in a list
#data = {"Route Name": route_names, "Route Link": route_links}

# Step 15: Convert the list to a DataFrame
df_route_names = pd.DataFrame(route_names)
df_route_names = df_route_names.assign(Links=route_links)

df_route_names.to_csv('APSRTC_Routes_Links.csv', index=False)

df_route_names
#route_links
```
